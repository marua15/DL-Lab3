```python
# step 1 : import modules
import requests
from bs4 import BeautifulSoup
import csv
from itertools import zip_longest
```

```python
# step 2 : use requests to fetch the url
result = requests.get("https://hihi2.com/")
```

```python
# step 3 : save page content/markup
src = result.content
print(src)
```

```python
#step 4 : create soup object to parse content
soup = BeautifulSoup(src, "lxml")
print(soup)
```

```python
# step 5 : find the elements containing infos we need
#article_title = soup.find_all("a", {"class" :"cA4208fxAK"})
# Find all elements with the specified class
elements_with_class = soup.find_all(class_='cA4208fxAK')

# Extract the text content of each element
article_titles = [element.text.strip() for element in elements_with_class]

# Print the article titles
for title in article_titles:
    print("Title:", title)

# Find all <p> elements
paragraphs = soup.find_all('p')

# Extract the text content of <p> elements
paragraph_texts = [p.text.strip() for p in paragraphs]

# Print the time information and paragraphs
for paragraph in zip(paragraph_texts):
    print("Paragraph:", paragraph)
```

```python
#step 6 : loop over returned lists to extract info needed in other lists
titles_and_paragraphs = []

for title, paragraph in zip(article_titles, paragraph_texts):
    titles_and_paragraphs.append({'Title': title, 'Paragraph': paragraph})

# Print the extracted data
for data in titles_and_paragraphs:
    print("Title:", data['Title'])
    print("Paragraph:", data['Paragraph'])
    print()
```

```python
# step 7 : create a csv file and fill it with values

# Define the file name
csv_file = 'extracted_data.csv'

# Define the header for the CSV file
header = ['Title', 'Paragraph']

# Write data to CSV file
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=header)

    # Write the header
    writer.writeheader()

    # Write the data
    for data in titles_and_paragraphs:
        writer.writerow(data)
```

```python
import csv

# Define the file name
csv_file = 'extracted_data.csv'

# Read data from CSV file
data = []
with open(csv_file, mode='r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for row in reader:
        title = row['Title']
        paragraph = row['Paragraph']
        # Calculate relevance score for each text
        relevance_score = (len(title) + len(paragraph)) / 2
        # Normalize relevance score to a scale of 0 to 10
        normalized_relevance_score = (relevance_score / max(len(title), len(paragraph))) * 10
        data.append({'Title': title, 'Paragraph': paragraph, 'RelevanceScore': normalized_relevance_score})

# Print each text's relevance score
for i, item in enumerate(data):
    print(f"Text {i+1} - Relevance Score: {item['RelevanceScore']:.2f}")

# Calculate the average relevance score for the whole data
total_relevance_score = sum(item['RelevanceScore'] for item in data)
average_relevance_score = total_relevance_score / len(data)

print("\nAverage Relevance Score for the Whole Data:", average_relevance_score)
```

```python
import csv
import nltk
from nltk import sent_tokenize, word_tokenize
from collections import Counter
import math

# Download NLTK resources (if not already downloaded)
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

# Define the file name
csv_file = 'extracted_data.csv'

# Read data from CSV file
data = []
with open(csv_file, mode='r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for row in reader:
        data.append({'Title': row['Title'], 'Paragraph': row['Paragraph']})

# Function to calculate Topic count
def calculate_topic_count(text):
    words = word_tokenize(text)
    tagged_words = nltk.pos_tag(words)
    noun_count = sum(1 for word, tag in tagged_words if tag.startswith('NN'))  # Count nouns
    return noun_count

# Function to calculate Topic Noun ratio
def calculate_topic_noun_ratio(text):
    words = word_tokenize(text)
    tagged_words = nltk.pos_tag(words)
    total_words = len(words)
    noun_count = sum(1 for word, tag in tagged_words if tag.startswith('NN'))  # Count nouns
    if total_words == 0:
        return 0
    return noun_count / total_words

# Function to calculate Topic sentence ratio
def calculate_topic_sentence_ratio(text):
    sentences = sent_tokenize(text)
    total_sentences = len(sentences)
    if total_sentences == 0:
        return 0
    topic_sentences = sum(1 for sentence in sentences if calculate_topic_count(sentence) > 0)
    return topic_sentences / total_sentences

# Calculate relevance score for each text
for i, item in enumerate(data):
    topic_count = calculate_topic_count(item['Title'] + ' ' + item['Paragraph'])
    topic_noun_ratio = calculate_topic_noun_ratio(item['Title'] + ' ' + item['Paragraph'])
    topic_sentence_ratio = calculate_topic_sentence_ratio(item['Title'] + ' ' + item['Paragraph'])
    relevance_score = math.log10(topic_count * topic_noun_ratio * topic_sentence_ratio * 10**5)
    # Ensure relevance score is between 0 and 10
    relevance_score = max(min(relevance_score, 10), 0)
    data[i]['RelevanceScore'] = relevance_score

# Print each text's relevance score
for i, item in enumerate(data):
    print(f"Text {i+1} - Relevance Score: {item['RelevanceScore']:.2f}")

# Calculate the average relevance score for the whole data
total_relevance_score = sum(item['RelevanceScore'] for item in data)
average_relevance_score = total_relevance_score / len(data)

print("\nAverage Relevance Score for the Whole Data:", average_relevance_score)
```
